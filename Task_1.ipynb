<div><p style="font-size:18px;"><h2>Summary</h2>
The Iris dataset is one of the most famous datasets in machine learning, commonly used for classification problems.
<ul>
<li>Total samples: 150 flowers</li>
<li>Total features: 4 input features(sepal_length, sepal_width, petal_length, petal_width)</li>
<li>Target variable: Species(Setosa, Versicolor, Virginica)</li>
<li>Number of classes: 3 species</li>
Each species has: <li>50 samples</li></p>
</div>

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load dataset
Iris = sns.load_dataset('iris')
# Display dataset structure
print("Shape of Dataset:")
print(Iris.shape)
print("\nColumn Names:")
print(Iris.columns)
print("\nFirst 5 Rows:")
print(Iris.head())
print("\nBasic Statistics:")
print(Iris.describe())
```

<div style="font-size:18px;"><h2>Scatter plots </h2>
<p>The following scatter plots show the relationship between species and it is clearly visible that</p><ul><li>Setosa is clearly seperated</li>
<li>Other species overlap slightly</li></ul>
<p>In pair plot Setosa is clearly seperated in almost all plots with very distinct cluster and versicolor & virginica some overlap. It is harder to seperate using sepal features but petal lenth vs petal width shows perfect clustering.

</div>


```python
# Sepal Length vs Sepal Width
sns.scatterplot(
    x='sepal_length',
    y='sepal_width',
    hue='species',
    data=   Iris
)

plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# Petal Length vs Petal Width
sns.scatterplot(
    x='petal_length',
    y='petal_width',
    hue='species',
    data=Iris
)

plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
#pair plot for feature importance intuition and clustering patterns
sns.pairplot(Iris, hue='species')

plt.show()
```
